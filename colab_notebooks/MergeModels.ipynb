```python
import json
from peft import PeftModel, PeftConfig
from transformers import AutoModelForCausalLM, AutoTokenizer
from huggingface_hub import login

# Montar directorio de drive
from google.colab import drive
drive.mount('/content/drive')
```

LLAMA

```python
# Ruta del archivo
adapter_config_path = "/content/drive/MyDrive/Colab Notebooks/ft_llama_model/final_model/llama-1b_prompt0_60_PreprocesadoES/adapter_config.json"

# Cargar, limpiar y guardar
with open(adapter_config_path, "r") as f:
    config = json.load(f)

# Eliminar el campo problemático
if "use_rslora" in config:
    del config["use_rslora"]

# Sobrescribir el archivo limpio
with open(adapter_config_path, "w") as f:
    json.dump(config, f, indent=4)
```

```python
# Ruta del modelo base y del adaptador LoRA
base_model_id = "meta-llama/Llama-3.2-1B-Instruct"
adapter_path = "/content/drive/MyDrive/Colab Notebooks/ft_llama_model/final_model/llama-1b_prompt0_60_PreprocesadoES"  # cambia esto

# Montar directorio de drive
from google.colab import drive
drive.mount('/content/drive')

# Token de Hugging Face
hf_token = "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"
login(hf_token)

# 1. Cargar el modelo base
model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    device_map="auto",
    torch_dtype="auto"
)

# 2. Cargar el adaptador sobre el modelo base
model = PeftModel.from_pretrained(model, adapter_path)

# 3. Fusionar manualmente los pesos del adaptador con el modelo base
model = model.merge_and_unload()

# 4. Guardar el modelo final completo (sin LoRA, ya fusionado)
output_path = "/content/drive/MyDrive/Colab Notebooks/ft_llama_model/final_model/llama_merged"
model.save_pretrained(output_path)

tokenizer = AutoTokenizer.from_pretrained(base_model_id)
tokenizer.save_pretrained(output_path)
```
